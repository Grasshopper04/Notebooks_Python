```python
import time
#goldbach for even real numbers greater than 2 as sum of two primes
def goldbach_conj_even(x,y):
    start = time.time()
    if x<y:
        prime=[]
        val=int(y)
        for i in range(0,val+1):
            if i>1:
                for l in range(2,i):
                    if i%l==0:
                        break
                else:
                    prime.append(i)
        even=[]
        for i in range(x,val+1):
            if i>0:
                if i%2==0:
                    even.append(i)
        primecomb=[]
        for i in prime:
            for j in prime:
                if i+j not in primecomb:
                    primecomb.append(i+j)
        primecomb.sort()
        for i in even:
            if i>2:
                if i in primecomb:
                    print(f'for the number {i} holds true')
                else:
                    print(f'fails for {i}')
    print(f'Time: {time.time() - start}')
goldbach_conj_even(0,20)

#goldbach for all real numbers greater than 2 as sum of three primes

def goldbach_conj_real(x,y):
    start = time.time()
    if x<y:
        prime=[]
        val=int(y)
        for i in range(0,val+1):
            if i>1:
                for l in range(2,i):
                    if i%l==0:
                        break
                else:
                    prime.append(i)
        real=[]
        for q in range(x,val+1):
            if q>1:
                real.append(i)
        primecomb=[]
        for i in prime:
            for j in prime:
                if i+j not in primecomb:
                    primecomb.append(i+j)
        primecomb.sort()
        for i in real:
            if i>2:
                if i in primecomb:
                    print(f'for the number {i} holds true')
                else:
                    print(f'fails for {i}')
    print(f'Time: {time.time() - start}')
goldbach_conj_real(0,20)
```
