```python
df['Male'] = np.where(df['Gender']=='Male', 1, 0)
df.head()
```

```python
import statsmodels.api as sm
```

```python
result = sm.stats.ttest_ind(df[df['Gender']=='Male']['Sleep'], 
          df[df['Gender']=='Female']['Sleep'])
print(result),

print ('t-stat:', result[0], '\n',)
print ('p-value:', result[1],  '\n'),
print ('d.f.:', result[2])
```
